```python
def cosine_similarity(query_vector, vector):
    # Computes cosine similarity between two vectors

    eps = 1e-10
    dot_product = sum(q * v for q, v in zip(query_vector, vector))

    query_magnitude = math.sqrt(sum(val ** 2 for val in query_vector))
    vector_magnitude = math.sqrt(sum(val ** 2 for val in vector))
    magnitude = query_magnitude * vector_magnitude + eps

    if magnitude == 0:
        return 0

    cosine_similarity = round(dot_product / magnitude, 5)
    return cosine_similarity
```

```python
def __chi2_distance(self, queryVector, vector):
    # Computes chi-square distance between two vectors

    eps = 1e-10
    dists = list(map(lambda q, v: ((q - v) ** 2) / (q + v + eps), queryVector, vector))
    chi2_distance = round(0.5 * sum(dists), 5)
    return chi2_distance

    
```

```python
def calculate_SAD_distance(query_vector, vector):
    # Computes SAD distance between two vectors

    dists = []
    for q, v in zip(query_vector, vector):
        dists.append(abs(q - v))

    SAD_distance = sum(dists)
    return SAD_distance

    
```
